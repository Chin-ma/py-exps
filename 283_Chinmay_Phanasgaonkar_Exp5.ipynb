# a) Write a menu-driven python code to make a linked-list

```python
#creating node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
#creating linked list        
class LinkedList:
    def __init__(self):
        self.first = None
        self.last = None
        
    #printing list   
    def printList(self):
        node = self.first
        print(" ")
        print("    Printing List")
        print("______________________\n")
        
        llist = []
        
        while node != None:
            llist.append(node.data)
            node = node.next
        if llist:
            print("Linked List:", llist)
        else:
            print("Linked List is Empty!!!")
    
    #add data to linked list
    def add(self, data):
        new_node = Node(data)
        if self.first == None:
            self.first = new_node
        if self.last != None:
            self.last.next = new_node
        self.last = new_node
        
    def size(self):
        current = self.first
        count = 0
        
        while current != None:
            count += 1
            current = current.next
        print("Size of list is:", count)
        print(" ")
            
    #searching element
    def search(self, item):
        current = self.first
        search_item = item
        found = False
        count = 1
        
        while current != None and not found:
            if int(current.data) == int(search_item):
                found = True
                print("Search Node found at position: ", count)
            else:
                current = current.next
                count += 1
            if current == None and found == False:
                print("Search Node", search_item, "Not Found!!!")
                print(" ")
            
    #removing element
    def remove(self, item):
        current = self.first
        previous = None
        delete_item = item
        found = False
       
        while not found:
            if current.data == delete_item:
                found = True
            else:
                previous = current
                current = current.next
       
        if previous == None:
            self.last = current.next()
        else:
            previous.next = current.next
            print("Element Deleted")
            
            
myLList = LinkedList()

menu = {}

menu['1'] = "Add Node"
menu['2'] = "Search Node"
menu['3'] = "Delete Node"
menu['4'] = "Print Linked List"
menu['5'] = "Exit"

while True:
    options = list(menu.keys())
    options.sort()
    print(" ")
    print("    Menu")
    print("____________\n")
    
    for entry in options:
        print(entry, menu[entry])
      
    selection = input("Enter your choice: ")
    
    if selection == '1':
        print("\n    Adding Nodes")
        print("___________________\n")
        length = int(input("Enter Number of Nodes you want to insert: "))
        
        for counter in range(1, length+1):
            item = int(input("Enter the Linked List Element: "))
            myLList.add(item)
            
        myLList.printList()
        myLList.size()
        
    elif selection == '2':
        print("\n    Searching Node")
        print("______________________\n")
        search_element = int(input("Enter Element to be searched: "))
        myLList.search(search_element)
        
    elif selection == '3':
        print("\n    Deleting Node")
        print("_____________________\n")
        delete_element = int(input("Enter the Node Element to be deleted: "))
        myLList.remove(delete_element)
        myLList.printList()
        myLList.size()
        
    elif selection == '4':
        myLList.printList()
        myLList.size()
        
    elif selection == '5':
        break
        
    else:
        print("Unknown Option Selected!!!")
```

# b) Write a menu driven python code to implement a Stack

```python
class Stack:
    def __init__(self):
        self.stack = []
        
    def empty(self):
        return len(self.stack) == 0
    
    def push(self, element):
        self.element = element
        self.stack.append(self.element)
        
    def delete(self):
        if self.empty():
            print("\nStack is underflow No element can be popped!!")
        else:
            print("\nElement popped out of the stack is:", self.stack.pop())
            
    def peek(self):
        if self.empty():
            print("\nStack is underflow No element can be peeked!!")
        else:
            print("\nSize of the stack is:", len(self.stack))
            print("Stack elements are:", self.stack, "\n")
            
    def search(self, element):
        self.element = element
        if self.element in self.stack:
            print("\nElement", self.element,"Exist at index:", self.stack.index(self.element))
        else:
            print("\nElement", self.element, "doesn't exist in the stack!!")

st = Stack()

menu = {}

menu['1'] = "PUSH"
menu['2'] = "POP"
menu['3'] = "PEEK"
menu['4'] = "Search"
menu['5'] = "Exit"

while True:
    options = list(menu.keys())
    options.sort()
    print(" ")
    print("    Menu")
    print("____________\n")
    
    for entry in options:
        print(entry, menu[entry])
        
    selection = input("Enter Your choice: ")
    
    if selection == '1':
        element = int(input("Enter Element to be pushed to stack: "))
        st.push(element)
    elif selection == '2':
        st.delete()
    elif selection == '3':
        st.peek()
    elif selection == '4':
        element = int(input("Enter element to be searched: "))
        st.search(element)
    elif selection == '5':
        break
    else:
        print("\nInvalid choice!!")
```

# c) Write a menu driven program in python to implement Queue

```python
class Queue_operations:
    def __init__(self):
        self.queue = []
        
    def empty(self):
        return len(self.queue) == 0
    
    def enQueue(self, element):
        self.element = element
        self.queue.append(self.element)
        
    def deQueue(self):
        if self.empty():
            print("\nQueue is empty!!")
        else:
            print("\nElement dequeued out of the queue is: ", self.queue.pop(0))
            
    def front(self):
        if self.empty():
            print("\nQueue is empty!!")
        else:
            print("\nElement at front is:", self.queue[0])
            
    def search(self, element):
        self.element = element
        if self.element in self.queue:
            print("\nElement", self.element, "exist at:", self.queue.index(self.element))
        else:
            print("\nSearch element doesn't exist in the queue!!")
            
    def display(self):
        if self.empty():
            print("\nQueue is empty!!")
        else:
            print("\nSize of the queue is:", len(self.queue))
            print("Queue elements are:", self.queue)
            
qo = Queue_operations()

menu = {}

menu['1'] = "EnQueue"
menu['2'] = "DeQueue"
menu['3'] = "Front"
menu['4'] = "Search"
menu['5'] = "Display"
menu['6'] = "Exit"

while True:
    options = list(menu.keys())
    options.sort()
    print(" ")
    print("    Menu")
    print("____________\n")
    
    for entry in options:
        print(entry, menu[entry])
        
    selection = input("Enter Your Choice: ")
    
    if selection == '1':
        element = int(input("Enter Element to be enqueued: "))
        qo.enQueue(element)
    elif selection == '2':
        qo.deQueue()
    elif selection == '3':
        qo.front()
    elif selection == '4':
        element = int(input("Enter element to be searched: "))
        qo.search(element)
    elif selection == '5':
        qo.display()
    elif selection == '6':
        break
    else:
        print("\nInvalid Choice!!")
```
